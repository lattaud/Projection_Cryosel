```python
from DMES_Libraries import *

DME = DMES_Parameters()
DME.SetDMmass(1e7)
energies,rates = DME.GetDMLinesAndRates()
plt.plot(energies,rates)
```

There are 200 ALP masses that can be selected either via their index :

```python
DME.SetDMmassindex(40)
print(f" mass selected : {DME.GetDMmass()} eV")
```

or by directly asking the mass you want. In that case, the mass selected will be the closest available.

```python
DME.SetDMmass(2.6*1e6)
print(f" mass selected : {DME.GetDMmass()} eV")
```

Of course if you change the value of $\sigma$ (Cross Section [$cm^{2}$]), the rate given in /kg/day changes.

```python
DME.SetCrossSection(1e-37)
print(f" current cross section value : {DME.GetCrossSection()} $cm^{2}$")
print("mass selected = ",DME.GetDMmass())
```

For each of these masses, There are 500 rate values from 0 to 50 eV with 0.1 eV width that can be plotted with the method PlotSpectrum()

```python
DME.PlotSpectrum()
```

These values can also be exported as numpy arrays with GetDMLinesAndRates

```python
DME.SetCrossSection(1e-36)
print(f"current cross section value : {DME.GetCrossSection()} cm^2")
print("mass selected = ",DME.GetDMmass())
energies,rates = DME.GetDMLinesAndRates()
_ = plt.plot(energies,rates)
```

Just to give an idea of how the spectral shapes varies with mass : 

```python
indexes = [x for x in list(range(20,54)) if x%5==0]
for i in indexes:
    DME.SetDMmassindex(i)
    energies,rates = DME.GetDMLinesAndRates()
    plt.plot(energies,rates,label=f"index = {i}, mass = {DME.GetDMmass()/1e6:10.2f} MeV")
    plt.yscale("log")
    plt.xlabel("Energy [eV]")
    plt.ylabel("Events / kg / day / 0.1 eV")
    plt.ylim(bottom=1e-5,top=1e7)
    plt.legend()
```
